```python
#convert the sure match
# first trying for sure match
# lower case (done)
# short form match
# n-grams match (considering first word only)
# no space all-in-one

def n_grams(list_name):
    name_len=len(list_name)
    new_list=list()
    st=list_name[0]
    for i in range(name_len):
        new_list.append(st)
        if i<name_len-1:
            st=st+" "+list_name[i+1]
    return new_list

def exact_match_check(comp,comp_lis):
    #lower and upper case match
    #comp_lis=list(df[ms_feild].unique())
    comp=comp.title()
    for i in comp_lis:
        i=str(i)
        i=i.title()
        
        #first case
        if i==comp:
            return i
        
        #n-grams word match word exact match
        spl=i.split(" ")
        n_gram_list=n_grams(spl)
        
        if (comp in n_gram_list):
            return i
        else:
            return comp
```

```python
#calculate score and ask user to update if any changes is required
from Levenshtein import ratio
import math
temp_dataframe=pd.DataFrame(columns=['slave','suggestion','score'])
def calculate_score(comp,comp_lis,option):
    comp=comp.title()
    score_list=list()
    score_old=0
    for i in comp_lis:
        i=str(i)
        i=i.title()
        
        score=ratio(comp,i)
        score=round(score,2)
        score_list.append(score)
        
        if (math.isclose(score_old, score) or (score_old < score)):
            suggest=i
            score_old=score
        
    score_list.sort(reverse=True)
    
    
    if(score_list[0]!=1):
        entry = {'slave': comp, 'suggestion': suggest, 'score': score_list[0]}
        print(entry)
        return entry 

choice=0
a_list=list()
a_list.append(df2["Manufacture"].apply(calculate_score,comp_lis=uni_lis,option=choice))
for data in a_list:
    data = list(filter(lambda item: item is not None, data))
result=pd.DataFrame(data)
```
